```python
# Código para comparar g2(t) para térmica, coherente y sub-Poisson
import numpy as np
import matplotlib.pyplot as plt

# eje de retardos τ
tau = np.linspace(-5, 5, 2001)  # unidades arbitrarias (ej. microsegundos)

# parámetros ejemplo
tau_c = 0.5   # tiempo de coherencia para fuente térmica
tau_a = 0.3   # tiempo de relajación (antibunching) para fuente sub-Poisson
n_prom = 1.0  # no usado en fórmulas simples, solo referencia

# coherente
g2_coh = np.ones_like(tau)

# térmica (single-mode approx con g1 = exp(-|τ|/tau_c))
g2_thermal = 1.0 + np.exp(-2.0 * np.abs(tau) / tau_c)

# sub-Poisson / antibunched simple model (idealized two-level emitter, incoherent)
g2_sub = 1.0 - np.exp(-np.abs(tau) / tau_a)

# plot
plt.figure(figsize=(8,4))
plt.plot(tau, g2_thermal, label='Térmica (g2=1+e^{-2|τ|/τc})')
plt.plot(tau, g2_coh,    label='Coherente (g2=1)')
plt.plot(tau, g2_sub,    label='Sub-Poisson / Antibunching (g2=1-e^{-|τ|/τa})')
plt.axvline(0, color='gray', linestyle='--', linewidth=0.6)
plt.xlim(-1.5, 1.5)
plt.ylim(0, 2.2)
plt.xlabel('Retardo τ (unidades arbitrarias)')
plt.ylabel('g²(τ)')
plt.title('Comparación de g²(τ): térmica vs coherente vs sub-Poisson')
plt.legend()
plt.grid(True)
plt.show()
```
